```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
nflx = yf.download('NFLX',start='2020-01-01')['Adj Close']
nflx = nflx.pct_change()
plt.figure(figsize=(15,10))
plt.plot(nflx.rolling(30).std()*np.sqrt(252)*100)
```
